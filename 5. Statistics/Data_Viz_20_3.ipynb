```python
flight_data = pd.read_csv(r'C:\Users\maham\Documents\flight_delays.csv', index_col='Month')
```

```python
flight_data[:12]
```

```python
plt.figure(figsize=(16, 8))
plt.title('Average arrival delays for the ariline WN, by month')

#bar graph
sns.barplot(x=flight_data.index, y=flight_data['WN'])

#add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

# Heatmap


```python
plt.figure(figsize=(10, 8))
plt.title('Average arrival delays for the ariline WN, by month')

#Heatmap
sns.heatmap(data=flight_data, annot=True, cmap='BuGn')

#add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```

```python
sns.heatmap?
```

```python
flight_data.corr()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Average arrival delays for the ariline WN, by month')

#Heatmap
corr = flight_data.corr()
sns.heatmap(data=corr, annot=True, cmap='PRGn')

#add the labels
plt.xlabel('Months')
plt.ylabel('Arrival delays')
```
